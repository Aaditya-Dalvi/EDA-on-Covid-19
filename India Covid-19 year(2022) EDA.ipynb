# Exploratory Data Analysis on Covid-19 cases in India

### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

### 1. COVID-19 India Statewise Vaccine Data

```python
df = pd.read_csv('COVID-19 India Statewise Vaccine Data.csv')
df.drop(index=36,inplace=True)
df
```

```python
df.shape
```

#### Adding two columns 'population' and 'percentage completely vaccinated' for better analysis

```python
# A Completely Vaccinated Person Is One Who Has Administered Both Vaccines. Number of People under dose 2 refers that their 
# vaccination is completed.

df['Population']  = [380581,49576777,1383727,31205576,104099452,1055450,25545198,586956,16787941,1458545,60439692,25351462,6864602,12258433,32988134,61095297,33406061,274289,64473,72626809,112374333,2855794,2966889,1097206,1978502,41974218,1247953,27743338,68548437,610577,72147030,35004000,3673917,199812341,10086292,91276115]
df["Percentage Completely Vaccinated"]=(df['Dose 2']*100)/df["Population"]
df

# df.head(10).style.set_properties(**{'background-color':'white','color': 'purple','margin':'True'})
```

```python
df.info()
```

```python
df.describe(include='all').T
```

The overall mean of Percentage of Completely Vaccinated states is **12.65603**

Now, sorting values based on features that are to be analyzed


```python
df1 = df.sort_values(by='Total Vaccination Doses',ascending = False)
df2 = df.sort_values(by='Dose 2',ascending=False)
df3 = df.sort_values(by='Percentage Completely Vaccinated',ascending=False)
```

# EDA using visualizations




###  **EDA Of States With Most as well as Least Total Number Of Jabs(injected with vaccine) Given**

# Taking A Look At Covid-19 Cases Situation In India¶


```python
dx= pd.read_csv("Latest Covid-19 India Status.csv")
dx = dx.rename(columns={'State/UTs':'State'})
print(dx.shape)
dx.head()
```

Merging both datasets will allow for a comprehensive analysis of state-wise COVID-19 metrics, including **Total Cases, Deaths** and **Death Ratio** with their corresponding vaccination statuses.

```python
dxnew = pd.concat([dx,df],join='inner',axis=1)
dxnew = dxnew.drop(columns=['State'])
print(dxnew.shape)
dxnew.head()
```

```python
dxnew.info()
```

```python
dxnew.describe()
```

```python
dx1 = dxnew.sort_values(by=['Total Cases'],ascending=False)
dx2= dxnew.sort_values(by=['Deaths'],ascending=False)
dx3= dxnew.sort_values(by=['Death Ratio (%)'],ascending=False)
```

## State-wiseTotal Cases 

### Highest Total of Covid-19 Cases within a state

```python
dxsort1_High = dx1.iloc[0:8]
dxsort1_High
```

```python
ax = sns.catplot(x='State/UTs', y='Total Cases', data=dxsort1_High, kind='bar', aspect=2.5, height=4)
for p in ax.ax.patches:
    ax.ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('States With Highest Total cases')


ay = sns.catplot(x='State/UTs', y='Percentage Completely Vaccinated', data=dxsort1_High, kind='bar', aspect=2.5, height=4)
for p in ay.ax.patches:  
    ay.ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.title('Corresponding Percentage of Vaccination completed') 

plt.show()
```

```python
sns.catplot(x='State/UTs', y='Total Cases', data=dxsort1_High, kind='point', aspect=3.5, height=3,color='lightgreen')
plt.title('States With Highest Total cases')


sns.catplot(x='State/UTs', y='Percentage Completely Vaccinated', data=dxsort1_High, kind='point', aspect=3.3, height=3)
plt.title('Corresponding Percentage of Vaccination completed') 

plt.show()
```

### Lowest Total of Covid-19 Cases within a state

```python
dxsort1_Low = dx1.iloc[-1:-9:-1]
dxsort1_Low
```

```python
ax = sns.catplot(x='State/UTs', y='Total Cases', data=dxsort1_Low, kind='bar', aspect=2.5, height=4)
for p in ax.ax.patches:
    ax.ax.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('States With Lowest Total cases')
plt.xticks(rotation=25,ha='right')

ay = sns.catplot(x='State/UTs', y='Percentage Completely Vaccinated', data=dxsort1_Low, kind='bar', aspect=2.5, height=4)
for p in ay.ax.patches:  
    ay.ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')
plt.title('Corresponding Percentage of Vaccination completed') 
plt.xticks(rotation=25,ha='right')

plt.show()
```

```python
sns.catplot(x='State/UTs', y='Total Cases', data=dxsort1_Low, kind='point', aspect=3.5, height=3,color='lightgreen')
plt.title('States With Lowest Total cases')
plt.xticks(rotation=25,ha='right')


sns.catplot(x='State/UTs', y='Percentage Completely Vaccinated', data=dxsort1_Low, kind='point', aspect=3.3, height=3)
plt.title('Corresponding Percentage of Vaccination completed') 
plt.xticks(rotation=25,ha='right')

plt.show()
```
